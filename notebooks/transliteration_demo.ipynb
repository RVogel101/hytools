# Armenian Transliteration & IPA — Test Notebook

This notebook lets you try **Armenian → Latin**, **Latin → Armenian**, and **Armenian → IPA** for single words, sentences, and full text. The module supports **Western**, **Eastern**, and **Classical** dialects.

**Yes, it is built for sentences and full text:** the same functions process any string (word, sentence, or paragraph); non-Armenian characters are left unchanged.

Run from project root: `jupyter notebook notebooks/transliteration_demo.ipynb`

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd()))

from linguistics.transliteration import (
    to_latin,
    to_armenian,
    to_ipa,
    format_wa_latin_sentence,
    get_armenian_to_latin_map,
    get_armenian_to_ipa_map,
)
from linguistics.transliteration import Dialect
```

## 1. Single words — Western (default)

```python
words_wa = ["պետք", "ժամ", "ջուր", "ոչ", "իւր", "յոյս", "տուն", "սիրել", "տիեզերական", "արժէքաւոր", "յանկարծական", "գուլպայ", "օտարական", "փոզոց", "ենթակայ", "լուրջ","համ", "շողարծակում"]
for w in words_wa:
    lat = to_latin(w, "western")
    ipa = to_ipa(w, "western")
    print(f"  {w}  →  Latin: {lat}  |  IPA: {ipa}")
```

## 2. Dialect comparison (same word)

```python
word = "ջուր"  # water
for d in ("western", "eastern", "classical"):
    print(f"  {d:10}  Latin: {to_latin(word, d):12}  IPA: {to_ipa(word, d)}")
```

## 3. Latin → Armenian (reverse)

```python
# Western: "ou" / "oo" → ու, "u" → ը
for roman in ["bedk'", "choor", "vochʼ", "yoor", "hoys", "doun"]:
    arm = to_armenian(roman, "western")
    print(f"  {roman}  →  {arm}")
```

## 4. Round-trip (Armenian → Latin → Armenian)

```python
for w in ["պետք", "ժամ", "ջուր", "ոչ", "մնալ", "ոռ", "սպասել", ""]:
    lat = to_latin(w, "western")
    back = to_armenian(lat, "western")
    ok = "✓" if back == w else "≠"
    print(f"  {w} → {lat} → {back}  {ok}")
```

## 5. Sentences and full text

```python
sentence = "Ես կը սիրեմ հայերէն։"  # I love Armenian.
lat = to_latin(sentence, "western")
print("Armenian:", sentence)
print("Latin (WA):", lat)
print("Formatted:", format_wa_latin_sentence(lat))
print("IPA (WA):", to_ipa(sentence, "western"))
```

```python
# Longer passage (non-Armenian chars preserved)
text = "Տունը մեծ է։ Ան կը խօսի հայերէն։"
lat = to_latin(text, "western")
print("Raw Latin:", lat)
print("Formatted:", format_wa_latin_sentence(lat))  # Dounu medz e. An gu khosi hayeren.
```

## 7. Mapping tables (excerpt)

```python
wa_latin = get_armenian_to_latin_map("western")
print("Western Armenian → Latin (sample):")
for k in ["բ", "պ", "ճ", "ջ", "ը", "ո", "յ", "տ", "դ", "ձ", "ծ", "ղ", "չ", "շ", "ու", "իւ", "եա", "այ", "եյ", "ոյ"]:
    if k in wa_latin:
        print(f"  {k} → {wa_latin[k]}")
print("  (եա → 'ia' at end of surname; word-final յ in այ/եյ/ոյ silent unless 3-letter word)")
print()
wa_ipa = get_armenian_to_ipa_map("western")
print("Western Armenian → IPA (sample):")
for k in ["բ", "պ", "ճ", "ջ", "ը", "ո", "յ", "տ", "դ", "ձ", "ծ", "ղ", "չ", "շ", "ու", "իւ", "եա", "այ", "եյ", "ոյ"]:
    print(f"  {k} → {wa_ipa[k]}")
```

## 8. IPA output and round-trips (more examples)

```python
# Diverse vocabulary: Armenian → Latin and IPA (Western)
# Western uses insert_schwa by default (e.g. Ինքզինքս → inukʼuzinukʼus)
vocab = [
    "բայ", "հայ", "գոյ", "յոյս", "սիրել", "հայերէն", "ոչ", "ժամ", "ջուր",
    "պետք", "իւր", "տուն", "մեծ", "խօսիլ", "ան", "կը",
    "Զարթարեան", "Ինքզինքս", "սկսիլ", "սպասել", "բարեւ", "գեղեցիկ", "միայն",
]
print("Word        → Latin (WA)   → IPA (WA)")
print("-" * 55)
for w in vocab:
    lat = to_latin(w, "western")
    ipa = to_ipa(w, "western")
    print(f"  {w:14} → {lat:16} → {ipa}")
```

```python
# Round-trip: Armenian → Latin → Armenian (Western)
# Words with "թ" use tʼ in Latin so round-trip preserves թ (e.g. Զարթարեան → zarutʼarian)
for w in ["ժամ", "ջուր", "ոչ", "պետք", "հայերէն", "Զարթարեան", "Ինքզինքս"]:
    lat = to_latin(w, "western")
    back = to_armenian(lat, "western")
    ok = "✓" if back == w or back == w.lower() else "≠"
    print(f"  {w} → {lat} → {back}  {ok}")
```
