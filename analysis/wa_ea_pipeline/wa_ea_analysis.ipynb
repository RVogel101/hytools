# WA vs EA Interactive Analysis
This notebook loads `analysis/wa_ea_pipeline/wa_ea_pipeline_summary.json` and renders interactive Plotly visualizations.
Run the pipeline first (see `scripts/wa_ea_pipeline.py`).

```python
import json
from pathlib import Path
import plotly.express as px
import pandas as pd

summary_path = Path('analysis/wa_ea_pipeline/wa_ea_pipeline_summary.json')
if not summary_path.exists():
    print('Run the pipeline to create', summary_path)
else:
    data = json.loads(summary_path.read_text(encoding='utf-8'))
    pairs = data.get('pairs', [])
    rows = []
    flat = []
    for p in pairs:
        pi = p.get('pair_index')
        phon_mean = p.get('phonetic', {}).get('mean')
        ortho_mean = p.get('orthographic', {}).get('mean')
        rows.append({'pair_index': pi, 'phonetic_mean': phon_mean, 'orthographic_mean': ortho_mean})
        for v in p.get('phonetic_list', []):
            flat.append({'pair_index': pi, 'phonetic_distance': v})

    df = pd.DataFrame(rows)
    dff = pd.DataFrame(flat)

    if not df.empty:
        fig = px.scatter(df, x='phonetic_mean', y='orthographic_mean', hover_data=['pair_index'], title='Phonetic vs Orthographic (per pair)')
        fig.show()

    if not dff.empty:
        fig2 = px.box(dff, x='pair_index', y='phonetic_distance', title='Per-sentence phonetic distances by pair')
        fig2.show()
```
