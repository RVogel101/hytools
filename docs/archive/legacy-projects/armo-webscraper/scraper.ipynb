```python
# Parse text obtained
soup = BeautifulSoup(response.text, 'html.parser')

# Find all hyperlinks present on webpage
links = soup.find_all('a')

# From all links check for pdf link and
# if present download file
for link in links:
    if ('.pdf' in link.get('href', [])):

        # get pdf file link
        pdf_ref = link.get('href')
        
        # Get response object for link
        response = requests.get(pdf_ref)

        # parse file link to get file name
        file_name = os.path.basename(pdf_ref).split('/')[-1]

        # Write content in pdf file
        pdf = open(file_name, 'wb')
        pdf.write(response.content)
        pdf.close()
        
        print(file_name, ' downloaded')

print("All PDF files downloaded")
```

```python
# Parse text obtained
soup = BeautifulSoup(response.text, 'html.parser')

# Find all hyperlinks present on webpage
links = soup.find_all('a')
image_file_types = ['jpeg', 'jpg', 'png', 'tif','.gif','.tiff','svg']
# From all links check for pdf link and
# if present download file
for link in links:
    for image_file_type in image_file_types:
        if (image_file_type in link.get('href', [])):

            # get pdf file link
            image_ref = link.get('href')
            
            # Get response object for link
            response = requests.get(image_ref)

            # parse file link to get file name
            file_name = os.path.basename(image_ref).split('/')[-1]
            file_name = file_name.split('?', 1)[0]

            # Write content in pdf file
            image = open(file_name, 'wb')
            image.write(response.content)
            image.close()
            
            print(file_name, ' downloaded')

print("All PDF files downloaded")
```
